# 1.Write a Python program to check whether a given year is a leap year or not.


```python
year = int(input("Enter year: "))

if (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0):
    print("Leap Year")
else:
    print("Not a Leap Year")
```

# 2. Write a Python program to check whether a given character is a vowel or consonant. 

```python
ch = input("Enter a character: ").lower()

if ch in ['a', 'e', 'i', 'o', 'u']:
    print("Vowel")
else:
    print("Consonant")
```

# 3. Write a Python program to calculate grade of a student based on percentage.

```python
percentage = float(input("Enter percentage: "))

if percentage >= 90:
    print("Grade A")
elif percentage >= 75:
    print("Grade B")
elif percentage >= 60:
    print("Grade C")
elif percentage >= 40:
    print("Grade D")
else:
    print("Fail")
```

# 4. Write a Python program to check whether a given number is a palindrome or not. 

```python
num = int(input("Enter number: "))
temp = num
rev = 0

while num > 0:
    rev = rev * 10 + num % 10
    num //= 10

if temp == rev:
    print("Palindrome")
else:
    print("Not Palindrome")
```

# 5. Write a Python program to find whether a given number is divisible by both 5 and 11


```python
num = int(input("Enter number: "))

if num % 5 == 0 and num % 11 == 0:
    print("Divisible by 5 and 11")
else:
    print("Not divisible by 5 and 11")
```

# 6. Write a Python program to calculate electricity bill based on units consumed.

```python
units = int(input("Enter units: "))

if units <= 100:
    bill = units * 1.5
elif units <= 200:
    bill = 100 * 1.5 + (units - 100) * 2.5
else:
    bill = 100 * 1.5 + 100 * 2.5 + (units - 200) * 4

print("Electricity Bill:", bill)
```

# 7. Write a Python program to check whether a given number is a prime number. 

```python
num = int(input("Enter number: "))

if num > 1:
    for i in range(2, num):
        if num % i == 0:
            print("Not Prime")
            break
    else:
        print("Prime")
else:
    print("Not Prime")
```

# 8. Write a Python program to check whether a given character is an alphabet, digit, or special character. 

```python
ch = input("Enter a character: ")

if ch.isalpha():
    print("Alphabet")
elif ch.isdigit():
    print("Digit")
else:
    print("Special Character")
```

# 9. Write a Python program to find the roots of a quadratic equation.

```python
import math

a = float(input("Enter a: "))
b = float(input("Enter b: "))
c = float(input("Enter c: "))

d = b*b - 4*a*c

if d > 0:
    root1 = (-b + math.sqrt(d)) / (2*a)
    root2 = (-b - math.sqrt(d)) / (2*a)
    print("Roots are:", root1, root2)
elif d == 0:
    root = -b / (2*a)
    print("Roots are equal:", root)
else:
    print("No real roots")
```

# 10. Write a Python program to calculate bonus of an employee based on salary.

# 11. Write a Python program to determine the type of triangle (equilateral, isosceles, or scalene). 


```python
a = int(input("Enter side a: "))
b = int(input("Enter side b: "))
c = int(input("Enter side c: "))

if a == b == c:
    print("Equilateral Triangle")
elif a == b or b == c or a == c:
    print("Isosceles Triangle")
else:
    print("Scalene Triangle")
```

# 12. Write a Python program to calculate attendance eligibility for exam.

```python
total = int(input("Enter total classes: "))
attended = int(input("Enter attended classes: "))

percentage = (attended / total) * 100

if percentage >= 75:
    print("Eligible for Exam")
else:
    print("Not Eligible for Exam")
```
