```python
#multiple methods working together

class Student:
    def __init__(self,name):
        self.name=name
        self.marks=[]
        
    def add_mark(self,mark):
        self.marks.append(mark)
        
    def average(self):
        return sum(self.marks)/len(self.marks)
    
    def display(self):
        print("Name:",self.name)
        print("Marks:",self.marks)
        print("Average:",self.average())

s1=Student ("Amit")
s1.add_mark(80)
s1.add_mark(98)
s1.add_mark(79)

s1.display()
```

# Inheritance

```python
Inheritance allow a clas to reuse properties and method from another class

parent class (base class)-->original class that has properties and methods

child class (derived  class)-->new class that inherits from the parent class

it is used to avoid code duplication and organize related classes
```

```python
#create parent class

class Student:
    def __init__(self,first_name, last_name):
        self.first_name=first_name
        self.last_name=last_name
        
    def display(self):
        print(self.first_name,self.last_name)

s1=Student("Amit","Raj")
s1.display()
```

```python
#parent class
class Student:
    def __init__(self,first_name, last_name):
        self.first_name=first_name
        self.last_name=last_name
        
    def display(self):
        print(self.first_name,self.last_name)

#child class
class hsstudent(Student): #this child class inherits all methods and properties from Student
    pass

s2=hsstudent("Raj","SK")
s2.display()
```

```python
#parent class
class Student:
    def __init__(self,first_name, last_name):
        self.first_name=first_name
        self.last_name=last_name
        
    def display(self):
        print(self.first_name,self.last_name)

#child class
class hsstudent(Student):
    def __init__(self,first_name, last_name,grade):
        Student.__init__(self,first_name, last_name) # call parent constructor
        self.grade=grade
        
    def welcome(self):
        print(f"welcome {self.first_name} {self.last_name} to grade {self.grade}")
        
s2=hsstudent("Raj","SK",10)
s2.display()
s2.welcome()
```

```python
#parent class
class Student:
    def __init__(self,first_name, last_name):
        self.first_name=first_name
        self.last_name=last_name
        
    def display(self):
        print(self.first_name,self.last_name)

#child class
class hsstudent(Student):
    def __init__(self,first_name, last_name,grade):
        super().__init__(first_name, last_name) # using super()
        self.grade=grade
        
    def welcome(self):
        print(f"welcome {self.first_name} {self.last_name} to grade {self.grade}")
        
s2=hsstudent("Raj","SK",10)
s2.display()
s2.welcome()
```

```python
#different classes with same method

class hsstudent:
    def activity(self):
        print("Studying for exam")

class csstudent:
    def activity(Self):
        print("Attending class")
        
class gsstudent:
    def activity(Self):
        print("Doing research")

students=[hsstudent(),csstudent(),gsstudent()]

for student in students:
    student.activity()
```

##### polymorphism with inheritance

```python
#create parent class

class Student:
    def __init__(self,name):
        self.name=name
    def activity(self):
        print(f"{self.name} is studying")
              
#create child class
class hsstudent(Student): 
    pass

class csstudent(Student):
    def activity(self):
        print(f"{self.name} is Attending class")
        
class gsstudent(Student):
    def activity(self):
        print(f"{self.name} is Doing research")

students=[hsstudent("KT"),csstudent("SK"),gsstudent("AK")]

for student in students:
    student.activity()              

```

```python
class Student:
    def __init__(self,name):
        self.name=name #public property
        self.__grade=0 #private property
        
    def set_grade(self,grade):
        if 0<=grade<=100:
            self.__grade=grade
        else:
            print("Grade must be between 0 to 100")
            
    def get_grade(self):
        return self.__grade
    
    def get_status(self):
        if self.__grade>=60:
            return "Passed"
        else:
            return "Failed"
        
s1=Student("KT")
s1.set_grade(85)
print(s1.get_grade())
print(s1.get_status())

#print(s1.__grade) this will cause error
s1.set_grade(150)
```

```python
#inner class without outer class access
class Student:
    def __init__(self,name):
        self.name=name 
        
    class Grades:
        def __init__(self):
            self.marks = []

        def add_mark(self, mark):
            self.marks.append(mark)

        def show_marks(self):
            print("Marks:", self.marks)

s1=Student("KT") #create a student object
g1=s1.Grades() #create inner class object

g1.add_mark(98)
g1.add_mark(89)
g1.show_marks()
```

```python
#inner class with outer class access
class Student:
    def __init__(self,name):
        self.name=name 
        
    class Grades:
        def __init__(self,student):
            self.student=student #store outer class instance
            self.marks = []

        def add_mark(self, mark):
            self.marks.append(mark)

        def show_marks(self):
            print(f"{self.student.name}'s Marks:{ self.marks}")

s1=Student("KT") #create a student object
g1=s1.Grades(s1) #pass outer instance

g1.add_mark(98)
g1.add_mark(89)
g1.show_marks()
```
