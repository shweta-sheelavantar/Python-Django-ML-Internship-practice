```python
l=[10,20,30,56,67,78,"KT","BLR","Tech"]
print(type(l))
print(len(l))
```

```python
print(l[0])
print(l[0:4])
print(l[-1])
```

```python
#change the value
l[0:2]=50,60
print(l)
```

```python
#using for loop
import time
for i in l:
    print(i,end=" ")
    time.sleep(0.5)
```

```python
#count specific value

print(l.count(60))
```

```python
#it will print the index of the first value
print(l.index("KT"))
```

```python
#add the list items in the last
l.append("100")
print(l)
```

```python
#add the list item in the specific index
l.insert(0,"YPR")
print(l)
```

```python
l.pop()
print(l)
```

```python
#remove the specific value
l.remove(50)
print(l)
```

```python
li=[8,9,10,5,6,5,2,3,4,67,54]
print("min:",min(li))
print("max:",max(li))
print("sorted:",sorted(li))
```

```python
#reveese the list
li=[8,9,10,5,6,5,2,3,4,67,54]
li.reverse()
print(li)
```

```python
#join the list
l1=[1,2,3]
l2=[4,5,6]
li=l1+l2
print(li)
```

```python
#using the loop
for i in l2:
    l1.append(i)
print(l1)
```

```python
#using extend
l2.extend(l1)
print(l2)
```
