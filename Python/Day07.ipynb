## Recursive function

###### A Recursive function is a function that calls itsself insideits own function untill a condition is nest

#####Base case ==> a condition the stop the recurstion
#####recursive call==> the function calls itself with a smaller value 

```python
def fact(n):
    if n==1:
        return 1
    else:
        return n*fact(n-1)
fact(5)
```

```python
def countdown(n):
    if n==0:
        print("finished")
    else:
        print(n)
        countdown(n-1)
countdown(10)
```

### lambda function
a lambda function is a way to create small , anonymous function
commmonly used with sorted() , filter(), map()


```python
#map()
nums=[1,2,3,4]
def squre(x):
    return x**2
result=map(squre, nums)
print(list(result))
 
```

```python
nums =[1,2,3,4]

squre=list(map(lambda x:x**2,nums))
print(squre)
```

```python
#filter()
nums=[1,2,3,4,5,6]

evens=list(filter(lambda x:x%2==0,nums))
print(evens)
```

```python
16. Write a Python function to count vowels in a string. 
```

```python
def count_vowels(s):
    count = 0
    for ch in s.lower():
        if ch in "aeiou":
            count += 1
    return count

print(count_vowels("Shweta"))
```

```python
17. Write a Python function to check whether a string is a palindrome. 
```

```python
def is_palindrome(s):
    if s == s[::-1]:
        return "Palindrome"
    else:
        return "Not Palindrome"

print(is_palindrome("madam"))
```

```python
def power(base, exp):
    result = 1
    for i in range(exp):
        result *= base
    return result

print(power(2, 3))
```

```python
20. Write a Python function to check whether a number is Armstrong. 
```

```python
def is_armstrong(num):
    temp = num
    sum = 0
    while temp > 0:
        digit = temp % 10
        sum += digit ** 3
        temp //= 10
    return sum == num

print(is_armstrong(153))
```

```python
21. Write a Python function to generate Fibonacci series up to n terms. 
```

```python
def fibonacci(n):
    a, b = 0, 1
    for i in range(n):
        print(a, end=" ")
        a, b = b, a + b

fibonacci(7)
```

```python
22. Write a Python function to calculate the average of numbers in a list. 
```

```python
def average(lst):
    total = 0
    for num in lst:
        total += num
    return total / len(lst)

print(average([10, 20, 30, 40]))
```

```python
24. Write a Python function to find the smallest number in a list. 
```

```python
def smallest(lst):
    small = lst[0]
    for num in lst:
        if num < small:
            small = num
    return small

print(smallest([5, 2, 8, 1, 9]))
```

```python
25. Write a Python function to count even and odd numbers in a list. 
```

```python
def count_even_odd(lst):
    even = 0
    odd = 0
    for num in lst:
        if num % 2 == 0:
            even += 1
        else:
            odd += 1
    return even, odd

print(count_even_odd([1, 2, 3, 4, 5, 6]))
```

```python
27. Write a Python function to find the GCD of two numbers. 
```

```python
def gcd(a, b):
    while b != 0:
        a, b = b, a % b
    return a

print(gcd(12, 18))
```

```python
28. Write a Python function to find the LCM of two numbers. 
```

```python
def lcm(a, b):
    greater = max(a, b)
    while True:
        if greater % a == 0 and greater % b == 0:
            return greater
        greater += 1

print(lcm(4, 6))
```

```python
29. Write a Python function to calculate the electricity bill using units 
consumed. 
```

```python
def electricity_bill(units):
    if units <= 100:
        return units * 1.5
    elif units <= 200:
        return 100 * 1.5 + (units - 100) * 2.5
    else:
        return 100 * 1.5 + 100 * 2.5 + (units - 200) * 4

print(electricity_bill(250))
```

```python
 30. Write a Python function to display all prime numbers within a given 
range
```

```python
def primes_in_range(start, end):
    for num in range(start, end + 1):
        if num > 1:
            for i in range(2, num):
                if num % i == 0:
                    break
            else:
                print(num)

primes_in_range(1, 50)
```
