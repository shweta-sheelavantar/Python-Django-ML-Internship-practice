```python
# student management system
students=[]

def add_students():
    name=input("enter the name:")
    age=input("Enter the age:")
    course=input("Enter the course:")
    
    student={
        "name":name,
        "age":age,
        "course":course
    }
    students.append(student)
def view_students():
    if len(students)==0:
        print("no student found")
        return
    print("Student List")
    for i,s in enumerate(students,start=1):   #i for index value and s for item of the list ( students details )
        
        print(f"students{i}")      # function formate {}it is placeholder =if { }used inside ()
        print(f"Name   :{s['name']}")
        print(f"Age   :{s['age']}")
        print(f"course   :{s['course']}")
        print("-----------------------------------")
    print()
    
while True:
    choice=input("Enter choice:1:Add student,2:view student,3:Exit")
    if choice=="1":
        add_students()
    elif choice=="2":
        view_students()
    elif choice=="3":
        print("thank you")
        break
    else:
        print("Invalid choice! TRY again")
```

```python
tell()==> shows current file pointer position
```

```python
f=open("abc.txt","r")
print(f.tell())
f.read(7)
print(f.tell())
f.close()
```

```python
Seek() ==> moves file pointer to a specific position
```

```python
f=open("abc.txt","r")
f.seek(10)
print(f.read())
f.close()
```

```python
f=open("abc.txt","r")

f.seek(0,2)
print(f.tell())
f.close()
```
