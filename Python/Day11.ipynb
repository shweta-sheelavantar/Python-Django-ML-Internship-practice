# simple Calculator using class

```python
class calci():
    def __init__(self,x,y):
        self.x=x
        self.y=y
    def add(self):
        print("add:",self.x+self.y)
    def sub(self):
        print("sub:",self.x-self.y)
    def mul(self):
        print("Mul:",self.x*self.y)
    def div(self):
        print("Div",self.x/self.y)
        
while True:
        a=int(input("Enter the first numbere :"))
        b=int(input ("Enter the second number:"))
        op=input("Enter the operators(+,-,*,/):")
        s1=calci(a,b)
        if op=="+":
            s1.add()
        elif op=="-":
            s1.sub()
        elif op=="*":
            s1.mul()
        elif op=="/":
            s1.div()
        else:
            print("Invalid operators")
        c=input("Do you want to continue(yes/no)")
        if c=="yes"  or c=="YES":
            continue
        elif c=="no" or c=="NO":
            break
        else:
            print("Invalid input")
            break
```

```python
class BankAccount():
    bank_name="State Bank of India"
    def __init__(self,acc_no,name,balance=0):
        self.acc_no=acc_no
        self.name=name
        self.balance=balance
    def deposite(self,amount):
        if amount>0:
            self.balance=self.balance+amount
            print("Amount deposited Sucessfully")
        else:
            print("Invalid Amount")
    def withdraw(self,amount):
        if amount<=self.balance:
            self.balance= self.balance-amount
            print("Withdrawal Sucessfull")
        else:
            print("Insuceffient balnce")
    def check_balance(self):
        print("Available Balance:",self.balance)
        
    def transfer(self,other_acc,amount):
        if amount<=self.balance:
            self.balance= self.balance-amount
            other_acc=other_acc+amount
            print("Amount Transfer Successfully")
        else:
            print("Insufficent Balance")
            
    def display(self):
        print("-----------Account Details----------")
        print("Bank name:",BankAccount.bank_name)
        print("Account Number:",self.acc_no)
        print("Name:",self.name)
        print("Balance:",self.balance)
        
        
s1=BankAccount(1001,"Harish",10000)
s2=BankAccount(1002,"Mahesh",5000)
while True:
    print("---Welcome To SBI Bank---------")
    print("\n 1.Deposite \n 2.withdraw \n 3.check balance \n 4.Amount Transform \n 5.Disply \n 6.exit")
    choice=int(input("Enter the choice:"))
    if choice==1:
        amt=int(input("Enter the amount:"))
        s1.deposite(amt)
    elif choice==2:
        amt=int(input("Enter the amount:"))
        s1.withdraw(amt)
    elif choice==3:
        s1.check_balance()
    elif choice==4:
        amt=int(input("Enter the amount to tranfer"))
        s1.transfer(s2,amt)
    elif choice==5:
        s1.display()
    elif choice==6:
        print("Thank you for the banking with us")
        break
    else:
        print("Invalid Choice")
```
