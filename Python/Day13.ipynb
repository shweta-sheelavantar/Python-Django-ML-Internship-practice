### Exception Handling

```python
Exception handling is the process of detecting and handling error that occurs program 
exception without crashing the program
why?
      to manage runtime errors
       makes program safe, stable and user  friendly
       prevent program failure
      
```

```python
try:
    #code
except
    #code that executes if error occurs
    
```

```python
x=int("hello")
print("convertion into int")
```

```python
try:
    x=int("hello")
except:
    print("some error happened")
```

```python
try:
    result=10/0
except ZeroDivisionError:
        print("Cannot divide by zero")
```

```python
try:
    x=int("abc")
except ValueError:
    print("Invalid number!")
```

```python
# multiple blocks
try:
    x=int("10")
    y=10/0
except ValueError: # python runs only the first matching one
    print("Conversion Error")
except ZeroDivisionError:
    print("Division Error")
```

```python
try:
    num= int(input("Enter a number:"))
    result=10/num
except(ValueError,ZeroDivisionError):
    print("Ivalid  input or division by zero")
```

```python
try:
    x=int("hello")
except Exception as error:
    print("Error Occured :",error)
    
```

```python
#finally block==>

try:
    x=int("abc")
except:
    print("Error happened")
finally:
    print("This will always run")
    
```

```python
# else block==> else runs only when no error occurs

try:
    x=int("10")
except:
    print("Error happened")
else:
    print("no error")
```

```python
import matplotlib.pyplot as plt
sem=[1,2,3,4,5,6]
per=[67,58,85,96,97,98]

plt.title("percentage vs Sem")
plt.xlabel("sem")
plt.ylabel("percentage")
plt.plot(sem,per,marker="o" ,ms=15,mec="black",mfc="g",ls="-.",color="r",lw=2) #"s","*","X","^","V","o"

#plt.grid(True)
plt.grid(color="green",ls="--",alpha=0.9)
plt.show()
```

```python
import matplotlib.pyplot as plt
sem=[1,2,3,4,5,6]
per=[67,58,85,96,97,98]

per1=[90,78,95,85,75,84]

plt.title("percentage vs Sem")
plt.xlabel("sem")
plt.ylabel("percentage")

plt.plot(sem,per,marker="o" ,ms=15,mec="black",mfc="g",ls="-.",color="r",lw=2) #"s","*","X","^","V","o"
plt.plot(sem,per1,marker="o" ,ms=15,mec="black",mfc="g",ls="--",color="green",lw=2) #"s","*","X","^","V","o"

#plt.grid(True)
plt.grid(color="green",ls="--",alpha=0.9)
plt.show()
```

# Subplot

```python
import matplotlib.pyplot as plt

sem=[1,2,3,4,5,6]
per=[86,85,54,96,94,87]

per1=[90,74,85,48,88,99]

plt.subplot(2,1,1)
plt.plot(sem,per1)
plt.title("First subplot")

plt.subplot(2,1,2)
plt.plot(sem,per1)
plt.title("second subplot")
```

```python
import matplotlib.pyplot as plt

sem=[1,2,3,4,5,6]
per=[86,85,54,96,94,87]

per1=[90,74,85,48,88,99]
per2=[84,85,78,88,97,84]

plt.subplot(1,3,1)
plt.plot(sem,per1,color="red")
plt.title("First subplot")

plt.subplot(1,3,2)
plt.plot(sem,per1)
plt.title("second subplot")

plt.subplot(1,3,3)
plt.plot(sem,per2,color="green")
plt.title("thrid subplot")
```

```python
import matplotlib.pyplot as plt

sem=[1,2,3,4,5,6]
per=[84,95,85,96,74,89]

explode=(0.2,0,0,0,0,0)
plt.pie(per,labels=sem,autopct="%1.1f%%",startangle=90,webgeprops={"edgecolor":"black"},explode=explode)
plt.show()
```

```python
import matplotlib.pyplot as plt
height=[150,160,170,180,190,200]
weight=[50,60,65,70,75,85]
age=[10,20,25,30,40,50]
plt.scatter(height,weight,
           c=age,# color
           s=[50,100,150,200,250,300],
           cmap="rainbow",
           marker="o",
           edgecolor="black",
           alpha=0.8
           )
plt.xlabel("height (cm)")
plt.ylabel("weight(kg)")
plt.title("height (cm)")
plt.colorbar(label="Age")
plt.show()
```
