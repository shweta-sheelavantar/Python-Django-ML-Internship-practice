### In python when you call a function , you can pass arguments in two ways:
    positional arguments-->based on order
    keyword Argument---->based on parameter name

```python
def student(name,age ,grade):
    print("Name:",name)
    print("Age:",age)
    print("Grade",grade)
    
```

```python
#calling with positional arguments
student("Kt",20,"A")
```

```python
#calling with keyword arguments
student(age=19,name="Kt",grade="A")
```

```python
### mixing the positional and keyword arguments
# positional arguments must come first

student("Kt",age=19,grade="A")

student(name="Kt",19,grade="A") #wrong
```

```python
def sum_two_numbers(a, b):
    return a + b
```

```python
# 2. Check even or odd
def is_even_or_odd(n):
    return "Even" if n % 2 == 0 else "Odd"
```

```python
# 3. Factorial of a number
def factorial(n):
    if n < 0:
        return None
    result = 1
    for i in range(1, n + 1):
        result *= i
    return result
```

```python
# 4. Check prime number
def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, int(n ** 0.5) + 1):
        if n % i == 0:
            return False
    return True
```

```python
# 5. Maximum of two numbers
def max_of_two(a, b):
    return a if a > b else b
```

```python
# 6. Maximum of three numbers
def max_of_three(a, b, c):
    return max(a, b, c)
```

```python
# 7. Simple interest
def simple_interest(principal, rate, time):
    return (principal * rate * time) / 100
```

```python
# 8. Area of a circle
def area_of_circle(radius):
    pi = 3.14159
    return pi * radius * radius
```

```python
# 9. Square and cube of a number
def square_and_cube(n):
    return n ** 2, n ** 3
```

```python
# 10. Check leap year
def is_leap_year(year):
    return (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0)
```

```python
# 11. Reverse a number
def reverse_number(n):
    return int(str(n)[::-1])
```

```python
# 12. Check palindrome number
def is_palindrome(n):
    return str(n) == str(n)[::-1]
```

```python
# 13. Sum of digits of a number
def sum_of_digits(n):
    return sum(int(digit) for digit in str(abs(n)))
```

```python
# 14. Celsius to Fahrenheit
def celsius_to_fahrenheit(celsius):
    return (celsius * 9 / 5) + 32
```

```python
# 15. Largest element in a list
def largest_in_list(lst):
    return max(lst)
```
