# Activity Day 05

1.Write a Python program to find the factorial of a number using a loop. 

```python
num = int(input("Enter a number: "))
fact = 1

for i in range(1, num + 1):
    fact = fact * i

print("Factorial:", fact)
```

2. Write a Python program to print the multiplication table of a given number. 

```python
num = int(input("Enter a number: "))

for i in range(1, 11):
    print(num, "x", i, "=", num * i)
```

```python
3. Write a Python program to count the number of digits in a given number. 
```

```python
num = int(input("Enter a number: "))
count = 0

while num > 0:
    count += 1
    num //= 10

print("Number of digits:", count)
```

```python
4. Write a Python program to reverse a number using a loop. 
```

```python
num = int(input("Enter a number: "))
rev = 0

while num > 0:
    rev = rev * 10 + num % 10
    num //= 10

print("Reversed number:", rev)
```

```python
5. Write a Python program to check whether a number is a palindrome using 
loops. 
```

```python
num = int(input("Enter number: "))
temp = num
rev = 0

while num > 0:
    rev = rev * 10 + num % 10
    num //= 10

if temp == rev:
    print("Palindrome")
else:
    print("Not Palindrome")
```

```python
8. Write a Python program to generate the Fibonacci series up to N terms. 
```

```python
n = int(input("Enter number of terms: "))
a, b = 0, 1

print(a, b, end=" ")

for i in range(2, n):
    c = a + b
    print(c, end=" ")
    a = b
    b = c
```

```python
9. Write a Python program to check whether a number is an Armstrong 
number using loops. 
```

```python
num = int(input("Enter number: "))
temp = num
sum = 0

while temp > 0:
    digit = temp % 10
    sum += digit ** 3
    temp //= 10

if sum == num:
    print("Armstrong Number")
else:
    print("Not Armstrong")
```

```python
11. Write a Python program to print all prime numbers between 1 and 100. 
```

```python
for num in range(2, 101):
    for i in range(2, num):
        if num % i == 0:
            break
    else:
        print(num)
```

```python
12. Write a Python program to find the GCD of two numbers using a loop. 
```

```python
a = int(input("Enter first number: "))
b = int(input("Enter second number: "))

gcd = 1
for i in range(1, min(a, b) + 1):
    if a % i == 0 and b % i == 0:
        gcd = i

print("GCD:", gcd)
```

```python
13. Write a Python program to find the LCM of two numbers using loops. 
```

```python
a = int(input("Enter first number: "))
b = int(input("Enter second number: "))

lcm = max(a, b)
while True:
    if lcm % a == 0 and lcm % b == 0:
        break
    lcm += 1

print("LCM:", lcm)
```
